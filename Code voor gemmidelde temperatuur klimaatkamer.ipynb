```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import os
from matplotlib.colors import Normalize

# Laad de data in
data = pd.read_excel('Data_Basis_model.xlsx')
data = data.drop(0, axis=0).reset_index(drop=True)
RCMH = data['RCMH']
SPH = data['SPH']
Lal = data['Lal']
Ral = data['Ral']
Lvl = data['Lvl']
Rvl = data['Rvl']
CaH = data['CaH']
CMH = data['CMH']
LCMH = data['LCMH']

# Sensorposities
x = [2, 1, 0.1, 1.9, 0.1, 1.9, 1, 1, 0.1]
y = [2, 4, 0.1, 0.1, 4, 4, 0.1, 2, 2]

# Zorg dat de directory bestaat
os.makedirs('images', exist_ok=True)

# Bereken de gemiddelde temperatuur per sensor
temperature_avg = [
    RCMH.mean(), SPH.mean(), Lal.mean(), Ral.mean(), 
    Lvl.mean(), Rvl.mean(), CaH.mean(), CMH.mean(), LCMH.mean()
]

# Plot de gemiddelde temperatuur
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot()

# Stel een lineaire schaal in
vmin, vmax = 8.5, 13.5  # Pas aan op basis van de dataset
norm = Normalize(vmin=vmin, vmax=vmax)

# Maak de scatterplot
sc = ax.scatter(x, y, c=temperature_avg, cmap='plasma', norm=norm, s=10000)

# Voeg een colorbar toe
cbar = plt.colorbar(sc, ax=ax, pad=0.1, ticks=np.linspace(vmin, vmax, 6))
cbar.set_label('Gemiddelde Temperatuur (C)', rotation=270, labelpad=15)

# Labels en titel
ax.set_xlabel('X (m)')
ax.set_ylabel('Y (m)')
ax.set_title('Gemiddelde Temperatuur')

# Sla het figuur op
plt.savefig(
    'images/average_temperature.png', 
    dpi='figure', 
    bbox_inches='tight', 
    pad_inches=0.1, 
    facecolor='white', 
    edgecolor='auto'
)
plt.show()
```
